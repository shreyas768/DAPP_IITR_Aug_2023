```python
employed_ppl = df[df['DAYS_EMPLOYED']<0]
employed_ppl['DAYS_BIRTH'] = -1*employed_ppl['DAYS_BIRTH']
employed_ppl['DAYS_EMPLOYED'] = -1*employed_ppl['DAYS_EMPLOYED']
employed_ppl
```

```python
df[(df['FLAG_OWN_CAR']=='Y') & (df['FLAG_OWN_REALTY']=='Y')]
```

```python
#days_employed vs days_bir
sns.scatterplot(data = employed_ppl, y = 'DAYS_EMPLOYED',x='DAYS_BIRTH', alpha = 0.008, linewidth = 0 )
```

```python
employed_ppl['age_in_years'] = employed_ppl['DAYS_BIRTH']/365.25
sns.histplot(data=employed_ppl,x='age_in_years', bins = 50, alpha=0.3, color = 'r')
plt.xlabel('Age in years')


#setting bins means you're undersmoothening vs oversmoothening your plot
```
