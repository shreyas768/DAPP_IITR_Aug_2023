```python
import seaborn as sns
```

```python
sns.get_dataset_names()
```

```python
df = sns.load_dataset('tips')
```

```python
df.head()
```

### distplot()

```python
df.shape
```

```python
sns.displot(df['total_bill'], bins = 30)
```

### jointplot (Bi-variate Data)


kind = 'scatter','kde' etc.

```python
sns.jointplot(data=df,x='total_bill',y='tip',kind = 'hex')
```

```python
sns.jointplot(data=df,x='total_bill',y='tip',kind = 'reg')
```

```python
sns.jointplot(x=df['total_bill'],y=df['tip'],kind = 'kde') #kernel density estimation
```

### pairplot

```python
sns.pairplot(df)
```

```python
sns.pairplot(df,hue='sex', palette='Set1')
```

```python
sns.rugplot(data=df,x='tip')
sns.kdeplot(df['tip'])
```
