```python
import numpy as np
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
df = pd.read_csv(r"C:\Users\shrey\Downloads\portilla_ml\05-Seaborn\dm_office_sales.csv")
```

```python
df.head()
```

### Countplot()

```python
plt.figure(figsize=(10,3))
sns.countplot(x='level of education', data=df)
```

```python
plt.figure(figsize=(10,3))
sns.countplot(x='division', data=df, hue ='level of education', palette='Paired' )
```

### barplot()

```python
plt.figure(figsize=(10,3))
sns.barplot(x='division',data=df, y = 'sales', estimator=np.mean)
```

```python
plt.figure(figsize=(10,4))
sns.barplot(x='division',data=df, y = 'sales', estimator=np.mean, hue = 'level of education')
plt.legend(bbox_to_anchor=(1.01,1))
```

# Distribution within categories

```python
df1 = pd.read_csv(r"C:\Users\shrey\Downloads\portilla_ml\05-Seaborn\StudentsPerformance.csv")
df1.head()
```

```python
df1.info()
```

### BoxPlot()

```python
plt.figure(figsize=(12,4))
sns.boxplot(x='parental level of education',y='math score',data=df1)
```

```python
plt.figure(figsize=(12,4))
sns.boxplot(x='parental level of education',y='math score',data=df1, hue = 'gender')
plt.legend(bbox_to_anchor=(1.01,1))
```

```python
plt.figure(figsize=(12,8))
sns.boxplot(y='parental level of education',x='math score',data=df1, hue = 'gender',orient='h')
plt.legend(bbox_to_anchor=(1.01,1))
plt.show()
```

```python
plt.figure(figsize=(12,4))
sns.boxplot(x='parental level of education',y='math score',data=df1, hue = 'gender', width = 0.5)
plt.legend(bbox_to_anchor=(1.01,1))
```

### Violinplot

```python
plt.figure(figsize=(12,6))
sns.violinplot(x='parental level of education',y='math score',data=df1)
```

```python
plt.figure(figsize=(12,6))
sns.violinplot(x='parental level of education',y='math score',data=df1, hue = 'gender')
```

```python
plt.figure(figsize=(12,6))
sns.violinplot(x='parental level of education',y='math score',data=df1, hue = 'gender', split = True)
```

```python
plt.figure(figsize=(12,6))
sns.violinplot(x='parental level of education',y='math score',data=df1, hue = 'gender', split = True, inner = None)
```

```python
plt.figure(figsize=(12,6))
sns.violinplot(x='parental level of education',y='math score',data=df1, hue = 'gender', split = True, inner = "quartile")
```

```python
plt.figure(figsize=(12,6))
sns.violinplot(x='parental level of education',y='math score',data=df1, hue = 'gender', split = True, inner = "stick")
```
