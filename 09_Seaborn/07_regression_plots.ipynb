```python
import seaborn as sns
```

```python
df = sns.load_dataset('tips')
df
```

#### lmplot()

It displays linear ML models

```python
sns.lmplot(x='total_bill',y='tip',data =df)
```

```python
sns.lmplot(x='total_bill',y='tip',data =df, hue = 'smoker')
```

```python
sns.lmplot(x='total_bill',y='tip',data =df, hue = 'smoker',
          markers=['*','v'])
```

#### Grid

```python
sns.lmplot(x='total_bill',y='tip',data =df, hue = 'smoker',
          markers=['*','v'], col ='time')
```

```python
sns.lmplot(x='total_bill',y='tip',data =df, hue = 'smoker',
          markers=['*','v'], col ='time',row='day', aspect=0.9)
```
