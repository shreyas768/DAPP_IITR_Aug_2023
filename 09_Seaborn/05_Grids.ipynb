```python
import numpy as np
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
df = pd.read_csv(r"C:\Users\shrey\Downloads\portilla_ml\05-Seaborn\StudentsPerformance.csv")
df.head()
```

### catplot()

```python
sns.catplot(x='gender',y='math score',data = df, kind = 'box')
```

```python
sns.catplot(x='gender',y='math score',data = df, kind = 'box',row = 'lunch')
```

```python
plt.figure(figsize=(12,6))
sns.catplot(x='gender',y='math score',data = df, kind = 'bar',row = 'lunch', col = 'test preparation course')
```

### Pairgrid

```python
pt = sns.PairGrid(df)
pt = pt.map_upper(sns.scatterplot)
pt = pt.map_diag(sns.barplot)
pt = pt.map_lower(sns.kdeplot, c = 'r')
```

### FacetGrid

```python
plt.figure(figsize=(15,6))
fg = sns.FacetGrid(data=df,row='race/ethnicity',col='gender')
fg = fg.map(plt.scatter,'math score','writing score')
fg.add_legend()
plt.subplots_adjust(hspace=0.4,wspace = 1)
```
