### Seaborn Parameters


#### hue

```python
plt.figure(figsize=(12,6))
sns.scatterplot(x='salary',y='sales',data=df, hue = 'level of education')
```

```python
plt.figure(figsize=(12,6))
sns.scatterplot(x='salary',y='sales',data=df, hue = 'training level', palette= 'colorblind')
```

#### size

```python
plt.figure(figsize=(12,6))
sns.scatterplot(x='salary',y='sales',data=df, size = 'training level')
```

```python
plt.figure(figsize=(12,6))
sns.scatterplot(x='salary',y='sales',data=df, size = 250)
```

#### alpha

```python
plt.figure(figsize=(12,6))
sns.scatterplot(x='salary',y='sales',data=df, alpha = 0.4)
```

#### style

```python
plt.figure(figsize=(12,6))
sns.scatterplot(x='salary',y='sales',data=df, style = 'level of education')
```

```python
plt.figure(figsize=(12,6))
sns.scatterplot(x='salary',y='sales',data=df, style = 'level of education', hue = 'training level')
plt.savefig('example_dapp_18_aug.png')
```
