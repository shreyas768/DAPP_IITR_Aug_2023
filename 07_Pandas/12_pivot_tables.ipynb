### pivot_table() : this would add aggregation functions

```python
df
```

```python
pd.pivot_table(data = df,index='Company',aggfunc=np.mean)[['Licenses','Sale Price']]
```

```python
df.groupby('Company').sum()[['Licenses','Sale Price']]
```

```python
pd.pivot_table(data=df,
              index = ['Account Manager','Contact'],
              values = ['Sale Price'] ,
              aggfunc= np.mean)
```

```python
#multiple aggregations
pd.pivot_table(data = df,
              index = ['Account Manager','Contact'],
              values = ['Sale Price'],
              aggfunc= [np.mean,np.max,np.min, np.sum])
```

```python
df[df['Contact'] == 'Cindy Phoner']
```

```python
#add columns
pd.pivot_table(data=df,
              index = ['Account Manager','Contact'],
              values = ['Sale Price','Licenses'],
              columns = ['Status'],
              aggfunc=[np.mean,np.max,np.min, np.sum],
              fill_value=0).transpose()
```

```python
df
```

```python
pd.pivot_table(data =df,
              index = ['Account Manager','Contact','Company','Product','Status'],
              values = ['Sale Price'],
              aggfunc=np.mean)
```

```python
#final 'ALL' with margins = True

pd.pivot_table(data=df,
              index = ['Account Manager','Contact','Product'],
              values = ['Sale Price','Licenses'],
              aggfunc= np.max,
              fill_value=0,
              margins=True)
```

```python
pd.pivot_table(data=df,
              index = ['Account Manager','Status'],
              values = ['Sale Price','Licenses'],
              aggfunc= np.sum,
              fill_value=0,
              margins=True)
```
