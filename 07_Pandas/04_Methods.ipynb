```python
def kind(price):
    if price < 15:
        return 'Snacks'
    elif price >= 15 and price < 30:
        return 'Meal'
    else:
        return 'Extensive Meal'
```

```python
df['kind'] = df['total_bill'].apply(kind)
```

```python
df
```

```python
df['final_bill'] = df['total_bill'].apply(lambda x:x*1.05)
```

```python
df
```

```python
def cust_quality(total_bill,tip):
    if tip/total_bill > 0.2:
        return 'Rich'
    else:
        return 'Poor'
```

```python
df['cust_quality'] = df[['total_bill','tip']].apply(lambda df: cust_quality(df['total_bill'],df['tip']),axis=1)
```
