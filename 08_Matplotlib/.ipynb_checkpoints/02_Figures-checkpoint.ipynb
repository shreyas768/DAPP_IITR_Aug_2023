```python
import matplotlib.pyplot as plt
import numpy as np
```

```python
a = np.linspace(0,10,11)
b = a **3
```

```python
a
```

```python
b
```

```python
x = np.arange(0,10)
y = 2*x
```

```python
fig = plt.figure()#blank canvas

axes = fig.add_axes([0,0,1,1]) 
#left,bottom,width,height

axes.plot(x,y)
plt.show()
```

```python
fig = plt.figure()#blank canvas

axes = fig.add_axes([0,0,1,1]) 
#left,bottom,width,height

axes.plot(a,b)
plt.show()
```

### Add another set of axes to my figure

```python
fig = plt.figure()#blank canvas

axes1 = fig.add_axes([0,0,1,1]) #Large Figure
axes2 = fig.add_axes([0.3,0.3,0.5,0.5]) #Small Figure

axes1.plot(a,b)
axes1.set_xlabel('X label')
axes1.set_ylabel('Y label')
axes1.set_title('title of figure 1')

axes2.plot(x,y)
axes2.set_title('title of figure 2')
```

```python
fig = plt.figure()#blank canvas

axes1 = fig.add_axes([0,0,1,1]) #Large Figure
axes2 = fig.add_axes([0.3,0.3,0.5,0.5]) #Small Figure

axes1.plot(a,b)
axes1.set_xlabel('X label')
axes1.set_ylabel('Y label')
axes1.set_title('title of Big Figure')

axes2.plot(x,y)
axes2.set_title('title of zoomed-in Small Figure')
axes2.set_xlim(2,8)
axes2.set_ylim(5,15)
```

```python
fig = plt.figure()#blank canvas

axes1 = fig.add_axes([0,0,1,1]) #Large Figure
axes2 = fig.add_axes([0.3,0.3,0.5,0.5]) #Small Figure
axes3 = fig.add_axes([1.2,0,1,1])

axes1.plot(a,b)
axes1.set_xlabel('X label')
axes1.set_ylabel('Y label')
axes1.set_title('title of Big Figure')

axes2.plot(x,y)
axes2.set_title('title of zoomed-in Small Figure')
axes2.set_xlim(2,8)
axes2.set_ylim(5,15)

axes3.plot(x,y)
```
