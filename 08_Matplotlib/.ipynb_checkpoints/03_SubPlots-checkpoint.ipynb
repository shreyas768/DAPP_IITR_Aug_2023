### plt.tight_layout() for preventing overlapping of plots

```python
fig, axes = plt.subplots(nrows=2,ncols=2)
axes[1,1].plot(a,b,c='m')
axes[0,1].plot(x,y, c = 'r')
plt.tight_layout()
```

```python
fig, axes = plt.subplots(nrows=2,ncols=2, figsize = (15,6))
axes[0,0].plot(a,b)
axes[0,0].set_title('[0,0] title')

axes[1,1].plot(x,y)

axes[0,1].plot(b,a)
axes[0,1].set_title('[0,1] title')

axes[1,0].plot(y,x)

fig.subplots_adjust(wspace=0.5,
    hspace=0.5,)

fig.suptitle('this is super title', fontsize=21)
fig.savefig('subplots_17aug_dapp.png',bbox_inches='tight',edgecolor='r')
```

### Legends

```python
fig = plt.figure()
ax = fig.add_axes([0,0,1,1])
ax.plot(x,x**2,'r.-',label='x**2')
ax.plot(x,x**3,'m--',label='x**3')
ax.legend(loc= 0)
#ax.legend(loc= (1,1.5))
```

```python
#RGB hex codes
fig = plt.figure()
ax = fig.add_axes([0,0,1,1])
ax.plot(x,x**2,c = '#DC143C',label='x**2', lw = 0.2, ls ='--', marker = '+', ms = 18 )
ax.plot(x,x**3,c = '#B8860B',label='x**3',lw=5,ls=':', marker = 's')
ax.plot(x,x**3-x**2,c = 'black',label='x**3-x**2',lw=2,ls='-.', marker = 'o', ms = 15)
ax.legend(loc= 0)
#ax.legend(loc= (1,1.5))
```

```python
fig, ax = plt.subplots(figsize=(12,6))
li = ax.plot(x,x**2,c = '#DC143C',label='x**2', lw = 3, ls ='--' )
print(type(li))
li[0].set_dashes([15,10]) #format: line length, space length
```
