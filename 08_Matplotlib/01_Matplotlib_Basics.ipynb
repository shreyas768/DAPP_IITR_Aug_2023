## Basic Matplotlib Commands

```python
plt.plot(a,b)
plt.title('Here you can put title of your figure')
plt.xlabel('Put X axis label here')
plt.ylabel('This would be Y axis label')
plt.show()
```

```python
#Editing your x and y label
plt.plot(a,b)
plt.title('Here you can put title of your figure')
plt.xlabel('Put X axis label here')
plt.ylabel('This would be Y axis label')
plt.xlim(4,8)
plt.ylim(3,16)
plt.show()
```

```python
#Exporting figure
plt.plot(a,b)
plt.title('Here you can put title of your figure')
plt.xlabel('Put X axis label here')
plt.ylabel('This would be Y axis label')
plt.xlim(4,8)
plt.ylim(3,16)
plt.savefig('matplotlib_example.png')
```
