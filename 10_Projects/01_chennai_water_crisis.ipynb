```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
chennai = pd.read_csv(r"C:\Users\shrey\Desktop\DAPP_Aug2023\DATA\chennai_reservoir_levels.csv")
chennai
```

```python
chennai.info()
```

```python
import datetime
```

```python
chennai['Date'] = pd.to_datetime(chennai['Date'],format = '%d-%m-%Y')
```

```python
chennai.info()
```

```python
chennai[['POONDI',
         'CHOLAVARAM',
         'REDHILLS',
         'CHEMBARAMBAKKAM']] = chennai[['POONDI',
                                        'CHOLAVARAM',
                                        'REDHILLS',
                                        'CHEMBARAMBAKKAM']].apply(pd.to_numeric)
```

```python
chennai
```

```python
chennai.info()
```

```python
chennai['Year'] = chennai['Date'].apply(lambda x:x.year)
chennai['Month'] = chennai['Date'].apply(lambda x:x.month)
```

```python
chennai['Total'] = chennai[['POONDI',
                                        'CHOLAVARAM',
                                        'REDHILLS',
                                        'CHEMBARAMBAKKAM']].sum(axis=1)
```

```python
chennai
```

```python
YearWise = chennai.groupby('Year', as_index=False)
```

```python
YearWise.mean()
```

```python
import warnings
warnings.simplefilter('ignore')
```

```python
plt.figure(figsize=(14,8))
sns.barplot(x = 'Year',y='Total', data= YearWise.mean())
plt.title('Mean Yearly water level in all four reservoir')
plt.show()
```

```python
YearWise.mean().plot(x='Year',y = ['POONDI',
                                        'CHOLAVARAM',
                                        'REDHILLS',
                                        'CHEMBARAMBAKKAM'],figsize = (20,10))
```

```python
chennai.info()
```

```python
chennai['Month'] = pd.Categorical(chennai['Month'])
```

```python
chennai.info()
```

```python
chennai['Year'] = chennai['Year'].astype('category')
```

```python
chennai.info()
```

```python
plt.figure(figsize=(25,8))
sns.barplot(x='Year',y='Total',hue='Month',data=chennai)
```

```python
feb = chennai[chennai['Month']==2]
plt.figure(figsize=(20,8))
sns.barplot(x='Year',y='Total',data = feb)
```

# Inferences: 

### Although chennai receive some rainfall in July, Aug, Sep due to SW monsoon but majority of rainfall happens in Oct to Dec due to NE monsoon
### From 2006 to 2010, NE monsoon has much more contribution than SW monsoon
### But in last few years, either both SW and NE are contributing equally or SW is contributing much more than NE
### Declining NE monsoon rainfall in recent year is in line with declining water level in 4 reservois in corresponding period (Eg. 2014)
